```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.DataFrame(np.random.randn(10,4), index=pd.date_range('1/1/2000', periods=10), columns=list('ABCD'))
df1.plot()
```

```python
# Bar plot
df2 = pd.DataFrame(np.random.randn(10, 4), columns=['a', 'b', 'c', 'd'])
df2.plot.bar()
```

```python
# to produce a stacked bar plot, pass stacked=True
df2.plot.bar(stacked=True)
```

```python
# to get horizontal bar plots, use barh method
df2.plot.barh(stacked=True)
```

```python
# Histograms
df3 = pd.DataFrame({'a': np.random.randn(1000)+ 1, 'b': np.random.randn(1000), 
                    'c': np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
df3.plot.hist(bins=20)
df3.plot.hist(bins=20, alpha=0.5)
```

```python
# to get horizontal histogram
df3.plot.hist(orientation='horizontal', bins=20)
```

```python
# to display grid
plt.figure()
# to plot different histograms for each column
df3.diff().hist(color='k', alpha=0.5, bins=20)
```

```python
# Box plots
df4 = pd.DataFrame(np.random.randn(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df4.plot.box()
```

```python
color = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')
df4.plot.box(color=color, sym='r+')
```

```python
# Horizontal and custom-positioned boxplot
df4.plot.box(vert=False, positions=[1, 4, 5, 6, 8])
```

```python
# boxplot
df5 = pd.DataFrame(np.random.rand(10,5))
plt.figure()
bp1 = df5.boxplot()
```

```python
# create a satisfied boxplot using the 'by' keyword argument to create groupings
df6 = pd.DataFrame(np.random.rand(10,2), columns=['col1', 'col2'])
df6['X'] = pd.Series(['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'])
plt.figure()
bp1 = df6.boxplot(by='X')
```

```python
# Area Plot
df7 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df7.plot.area()
```

```python
df7.plot.area(stacked=False)
```

```python
# Scatter Plot
df8 = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
df8.plot.scatter(x='a', y='b')
```

```python
# Pie chart
df9 = pd.DataFrame(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], columns=['x'])
df9.plot.pie(subplots=True)
```
